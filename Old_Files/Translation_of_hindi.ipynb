```
from googletrans import Translator
import pandas as pd
import numpy as np
from pandas import DataFrame
```

```
df = pd.read_csv('/content/drive/My Drive/Projects/FoodDataset/makedata4.csv',names=["RecipeName","Ingredients", "Prep", "Cook","Total","Servings","Cuisine","Type","Diet","Instructions"],index_col=False)
```

```
df
```

```
translator = Translator()
```

```
df2 = df.copy()
```

```
def translate_df_column(df,column_name):
  translated_list = []
  for item in df[column_name]:
    item = translator.translate(item)
    translated_item = item.text
    translated_list.append(translated_item)

  return translated_list
```

```
def add_to_df(df,column_name,list1):
  df[column_name]=list1
```

```
translated_recipename = translate_df_column(df2,"RecipeName")
add_to_df(df2,"TranslatedRecipeName",translated_recipename)

translated_ingredients = translate_df_column(df2,"Ingredients")
add_to_df(df2,"TranslatedIngredients",translated_ingredients)

translated_instructions = translate_df_column(df2,"Instructions")
add_to_df(df2,"TranslatedInstructions",translated_instructions)

```
